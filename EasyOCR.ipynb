```
def text_recognition(file_path, text_file_name):
  reader = easyocr.Reader(["ru"])
  result = reader.readtext(file_path, detail=0, paragraph=True)
  with open (text_file_name, "a") as file:
    for line in result:
      file.write(f"{line}\n")
    file.write(f"----")
```

```
def contrast(input_image_path, output_image_path):
   color_image = Image.open(input_image_path)
   enhancer = ImageEnhance.Contrast(color_image)
   im_output = enhancer.enhance(1.5)
   im_output.save(output_image_path)
```

```
def brightness(input_image_path, output_image_path):
   color_image = Image.open(input_image_path)
   enhancer = ImageEnhance.Brightness(color_image)
   im_output = enhancer.enhance(1.5)
   im_output.save(output_image_path)
```

```
def transformation_image(input_path, output_path):
  sharpness(input_path, output_path)
  contrast(output_path, output_path)
  brightness(output_path, output_path)
  black_and_white(output_path, output_path)
```

```
def drawing_borders(image, bounds, color='black', width=3):
    draw = ImageDraw.Draw(image)
    for bound in bounds:
        p0, p1, p2, p3 = bound[0]
        draw.line([*p0, *p1, *p2, *p3, *p0], fill=color, width=width)
    return image

transformation_image("/content/Pass/22.jpg", "/content/Pass/22_new.jpg")
img = Image.open("/content/Pass/22_new.jpg")

reader = easyocr.Reader(["ru"])
bounds = reader.readtext("/content/Pass/22_new.jpg")

print(bounds)

drawing_borders(img, bounds)
```
